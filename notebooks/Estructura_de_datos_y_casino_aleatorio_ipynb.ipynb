```python
import random
import collections
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def ejecutar_simulacion():
    # =============================================================================
    # SIMULACI√ìN DE DADOS: ESTRUCTURAS DE DATOS Y ESTAD√çSTICA
    # =============================================================================

    print("="*80)
    print("AN√ÅLISIS COMPLETO: LANZAMIENTO DE DOS DADOS")
    print("="*80)

    # Par√°metros de la simulaci√≥n - INPUT DEL USUARIO
    while True:
        try:
            N_LANZAMIENTOS = int(input("Ingrese el n√∫mero de lanzamientos para la simulaci√≥n (ej: 1000, 100000): "))
            if N_LANZAMIENTOS > 0:
                break
            else:
                print("Por favor ingrese un n√∫mero positivo.")
        except ValueError:
            print("Por favor ingrese un n√∫mero entero v√°lido.")

    # =============================================================================
    # 1. ESTRUCTURAS DE DATOS B√ÅSICAS
    # =============================================================================
    print("\n1. ESTRUCTURAS DE DATOS UTILIZADAS:")
    print("-" * 40)

    # Lista para almacenar resultados (estructura de datos lineal)
    resultados = []
    lanzamientos_individuales = []  # Lista de tuplas para cada lanzamiento

    print(f"‚Ä¢ LISTA (resultados): Almacena {N_LANZAMIENTOS:,} sumas.")
    print(f"‚Ä¢ LISTA DE TUPLAS: Almacena cada lanzamiento como (dado1, dado2, suma).")

    # Simulaci√≥n con registro detallado
    print("\nGenerando datos...")
    for i in range(N_LANZAMIENTOS):
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        suma = dado1 + dado2

        # Almacenar en diferentes estructuras
        resultados.append(suma)  # Lista simple
        lanzamientos_individuales.append((dado1, dado2, suma))  # Lista de tuplas # [(1,2,5), (1,2,6)]

        # Mostrar los primeros 10 lanzamientos
        if i < 10:
            print(f"  Lanzamiento {i+1:2d}: Dado1={dado1}, Dado2={dado2}, Suma={suma}")

    if N_LANZAMIENTOS > 10:
        print(f"... (continuando hasta {N_LANZAMIENTOS:,} lanzamientos)")

    # =============================================================================
    # 2. COUNTER - ESTRUCTURA DE DATOS ESPECIALIZADA
    # =============================================================================
    print(f"\n2. COUNTER (collections.Counter):")
    print("-" * 40)

    # Counter es un diccionario especializado para contar [6,5, 2, 6]  2 1  5 1 6 2
    conteo = collections.Counter(resultados)
    print(f"‚Ä¢ Counter es una subclase de dict optimizada para el conteo de frecuencias.")
    print(f"‚Ä¢ Tipo de dato: {type(conteo)}")

    print("\nFrecuencias emp√≠ricas encontradas:")
    for suma in sorted(conteo.keys()):
        print(f"  Suma {suma:2d}: {conteo[suma]:,} veces")

    # M√©todos √∫tiles del Counter
    print(f"\n‚Ä¢ Suma m√°s com√∫n: {conteo.most_common(1)[0]}")
    print(f"‚Ä¢ Total de elementos contados: {sum(conteo.values()):,}")

    # =============================================================================
    # 3. ESTAD√çSTICAS DESCRIPTIVAS
    # =============================================================================
    print(f"\n3. ESTAD√çSTICAS DESCRIPTIVAS:")
    print("-" * 40)

    # Convertir a array de NumPy para c√°lculos estad√≠sticos eficientes
    datos_np = np.array(resultados)# [6,2,6] 14/3

    # Medidas de tendencia central
    media = np.mean(datos_np)
    mediana = np.median(datos_np)
    moda_resultado = stats.mode(datos_np)
    moda_valor = moda_resultado.mode
    moda_conteo = moda_resultado.count

    print(f"‚Ä¢ Media (promedio emp√≠rico): {media:.4f} (Te√≥rica: 7.0)")
    print(f"‚Ä¢ Mediana (valor central): {mediana:.4f}")
    print(f"‚Ä¢ Moda (valor m√°s frecuente): {moda_valor} (apareci√≥ {moda_conteo:,} veces)")

    # Medidas de dispersi√≥n
    varianza = np.var(datos_np)
    desviacion_std = np.std(datos_np)
    rango = np.max(datos_np) - np.min(datos_np)

    print(f"‚Ä¢ Varianza: {varianza:.4f} (Te√≥rica: ~5.833)")
    print(f"‚Ä¢ Desviaci√≥n est√°ndar: {desviacion_std:.4f} (Te√≥rica: ~2.415)")
    print(f"‚Ä¢ Rango: {rango} (de {np.min(datos_np)} a {np.max(datos_np)})")

    # =============================================================================
    # 4. PROBABILIDADES TE√ìRICAS VS EMP√çRICAS
    # =============================================================================
    print(f"\n4. AN√ÅLISIS DE PROBABILIDADES:")
    print("-" * 40)

    # Diccionario con probabilidades te√≥ricas (estructura clave-valor)
    prob_teorica = {s: len([(d1, d2) for d1 in range(1, 7) for d2 in range(1, 7) if d1 + d2 == s]) / 36 for s in range(2, 13)}

    print("SUMA ‚îÇ FRECUENCIA ‚îÇ PROB.EMP√çRICA ‚îÇ PROB.TE√ìRICA ‚îÇ DIFERENCIA ABS.")
    print("‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")

    diferencias_totales = 0
    for suma in range(2, 13):
        freq = conteo.get(suma, 0) # Usar .get para evitar errores si una suma nunca aparece
        prob_empirica = freq / N_LANZAMIENTOS
        prob_teorica_val = prob_teorica[suma]
        diferencia = abs(prob_empirica - prob_teorica_val)
        diferencias_totales += diferencia

        print(f"{suma:4d} ‚îÇ {freq:10,} ‚îÇ {prob_empirica:13.5f} ‚îÇ {prob_teorica_val:12.5f} ‚îÇ {diferencia:14.5f}")

    print(f"\n‚Ä¢ Diferencia promedio entre probabilidades: {diferencias_totales/11:.6f}")
    print("  (Este valor tiende a cero a medida que aumenta el n√∫mero de lanzamientos)")

    # =============================================================================
    # 5. VISUALIZACI√ìN AVANZADA
    # =============================================================================
    print(f"\n5. GENERANDO VISUALIZACIONES...")

    x = list(range(2, 13))
    y_empirica = [conteo.get(s, 0) for s in x]

    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f'An√°lisis Completo: {N_LANZAMIENTOS:,} Lanzamientos de Dos Dados', fontsize=16, fontweight='bold')

    # Gr√°fico 1: Distribuci√≥n de frecuencias vs Frecuencia Te√≥rica Esperada
    y_teorica_esperada = [prob_teorica[s] * N_LANZAMIENTOS for s in x]
    ax1.bar(x, y_empirica, color='royalblue', alpha=0.7, label='Frecuencia Emp√≠rica')
    ax1.plot(x, y_teorica_esperada, 'ro-', linewidth=2, markersize=6, label='Frecuencia Te√≥rica Esperada')
    ax1.set_title('Distribuci√≥n de Frecuencias (Emp√≠rica vs. Te√≥rica)')
    ax1.set_xlabel('Suma de los dados')
    ax1.set_ylabel('Frecuencia')
    ax1.legend()
    ax1.grid(axis='y', linestyle='--', alpha=0.6)

    # Gr√°fico 2: Comparaci√≥n de Probabilidades
    prob_emp = [f/N_LANZAMIENTOS for f in y_empirica]
    prob_teo = [prob_teorica[s] for s in x]
    ax2.bar([i-0.2 for i in x], prob_emp, width=0.4, color='lightcoral', label='Prob. Emp√≠rica')
    ax2.bar([i+0.2 for i in x], prob_teo, width=0.4, color='lightgreen', label='Prob. Te√≥rica')
    ax2.set_title('Comparaci√≥n de Probabilidades')
    ax2.set_xlabel('Suma de los dados')
    ax2.set_ylabel('Probabilidad')
    ax2.legend()
    ax2.grid(axis='y', linestyle='--', alpha=0.6)

    # Gr√°fico 3: Error Absoluto
    diferencias = [abs(p_emp - p_teo) for p_emp, p_teo in zip(prob_emp, prob_teo)]
    ax3.bar(x, diferencias, color='orange', alpha=0.8)
    ax3.set_title('Error Absoluto |P(Emp√≠rica) - P(Te√≥rica)|')
    ax3.set_xlabel('Suma de los dados')
    ax3.set_ylabel('Diferencia Absoluta')
    ax3.grid(axis='y', linestyle='--', alpha=0.6)

    # Gr√°fico 4: Distribuci√≥n Acumulativa
    cdf_empirica = np.cumsum(prob_emp)
    ax4.plot(x, cdf_empirica, 'b-', marker='o', label='CDF Emp√≠rica')
    ax4.set_title('Funci√≥n de Distribuci√≥n Acumulativa (CDF)')
    ax4.set_xlabel('Suma de los dados')
    ax4.set_ylabel('Probabilidad Acumulada')
    ax4.legend()
    ax4.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # =============================================================================
    # 6. AN√ÅLISIS DE CONVERGENCIA (LEY DE LOS GRANDES N√öMEROS)
    # =============================================================================
    print(f"\n6. AN√ÅLISIS DE CONVERGENCIA (Ley de los Grandes N√∫meros):")
    print("-" * 60)

    media_teorica = 7.0
    prob_7_teorica = 1/6

    # Generar puntos de muestra de forma logar√≠tmica para ver mejor la convergencia
    if N_LANZAMIENTOS < 100:
        tama√±os_muestra = np.unique(np.linspace(10, N_LANZAMIENTOS, 10, dtype=int))
    else:
        tama√±os_muestra = np.unique(np.logspace(1, np.log10(N_LANZAMIENTOS), num=50, dtype=int))
        if tama√±os_muestra[-1] != N_LANZAMIENTOS:
             tama√±os_muestra = np.append(tama√±os_muestra, N_LANZAMIENTOS)

    # Asegurarse de que no haya tama√±os de muestra cero
    tama√±os_muestra = tama√±os_muestra[tama√±os_muestra > 0]

    print(f"Analizando la convergencia en {len(tama√±os_muestra)} puntos de muestra (de 1 a {N_LANZAMIENTOS:,})...")

    medias_empiricas = [np.mean(datos_np[:n]) for n in tama√±os_muestra]
    probs_7_empiricas = [(np.sum(datos_np[:n] == 7) / n) for n in tama√±os_muestra]

    errores_media = [abs(m - media_teorica) for m in medias_empiricas]
    errores_prob7 = [abs(p - prob_7_teorica) for p in probs_7_empiricas]

    # Visualizaci√≥n de la convergencia
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    fig.suptitle('Convergencia hacia Valores Te√≥ricos', fontsize=16, fontweight='bold')

    ax1.plot(tama√±os_muestra, medias_empiricas, 'bo-', markersize=4, label='Media Emp√≠rica')
    ax1.axhline(y=media_teorica, color='r', linestyle='--', label=f'Media Te√≥rica ({media_teorica})')
    ax1.set_title('Convergencia de la Media')
    ax1.set_xlabel('N√∫mero de Lanzamientos')
    ax1.set_ylabel('Media de las Sumas')
    ax1.set_xscale('log')
    ax1.legend()
    ax1.grid(True, which="both", ls="--", alpha=0.6)

    ax2.plot(tama√±os_muestra, probs_7_empiricas, 'go-', markersize=4, label='P(Suma=7) Emp√≠rica')
    ax2.axhline(y=prob_7_teorica, color='r', linestyle='--', label=f'P(Suma=7) Te√≥rica ({prob_7_teorica:.4f})')
    ax2.set_title('Convergencia de la Probabilidad para la Suma = 7')
    ax2.set_xlabel('N√∫mero de Lanzamientos')
    ax2.set_ylabel('Probabilidad')
    ax2.set_xscale('log')
    ax2.legend()
    ax2.grid(True, which="both", ls="--", alpha=0.6)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    print(f"\nüìä AN√ÅLISIS FINAL DE CONVERGENCIA:")
    print("-" * 60)
    print(f"Tras {N_LANZAMIENTOS:,} lanzamientos:")
    print(f"‚Ä¢ Media final: {medias_empiricas[-1]:.6f} (Error: {errores_media[-1]:.6f})")
    print(f"‚Ä¢ P(Suma=7) final: {probs_7_empiricas[-1]:.6f} (Error: {errores_prob7[-1]:.6f})")

    if len(tama√±os_muestra) > 10:
        # Calcular la tasa de convergencia usando los √∫ltimos puntos
        ultimos_errores = np.log(errores_media[-5:])
        ultimos_n = np.log(tama√±os_muestra[-5:])
        # La teor√≠a (Teorema del L√≠mite Central) sugiere que el error es proporcional a 1/sqrt(N)
        # Por lo tanto, log(error) ‚âà C - 0.5 * log(N). La pendiente deber√≠a ser cercana a -0.5.
        pendiente, _, _, _, _ = stats.linregress(ultimos_n, ultimos_errores)
        print(f"‚Ä¢ Tasa de convergencia del error (pendiente log-log): {pendiente:.4f} (Te√≥rica: -0.5)")

    # =============================================================================
    # 7. RESUMEN DE ESTRUCTURAS DE DATOS UTILIZADAS
    # =============================================================================
    print(f"\n7. RESUMEN DE ESTRUCTURAS DE DATOS:")
    print("="*50)
    print(f"‚Ä¢ LISTA (list): Almacenamiento secuencial de {N_LANZAMIENTOS:,} resultados.")
    print(f"‚Ä¢ TUPLA (tuple): Estructura inmutable para registrar cada lanzamiento (dado1, dado2, suma).")
    print(f"‚Ä¢ DICCIONARIO (dict): Pares clave-valor para las probabilidades te√≥ricas. Acceso O(1).")
    print(f"‚Ä¢ COUNTER (collections.Counter): Subclase de dict para conteo de frecuencias optimizado.")
    print(f"‚Ä¢ ARRAY DE NUMPY (np.array): Estructura optimizada para c√°lculo num√©rico y estad√≠stico masivo.")

    # =============================================================================
    # 8. CONCEPTOS ESTAD√çSTICOS DEMOSTRADOS
    # =============================================================================
    print(f"\n8. CONCEPTOS ESTAD√çSTICOS DEMOSTRADOS:")
    print("="*50)
    print("‚Ä¢ Distribuci√≥n de Frecuencias: C√≥mo se distribuyen los resultados observados.")
    print("‚Ä¢ Medidas de Tendencia Central: Media, mediana y moda de los datos.")
    print("‚Ä¢ Medidas de Dispersi√≥n: Varianza y desviaci√≥n est√°ndar que miden la volatilidad.")
    print("‚Ä¢ Probabilidad Te√≥rica vs. Emp√≠rica: Comparaci√≥n entre lo que la matem√°tica predice y lo que la simulaci√≥n muestra.")
    print("‚Ä¢ Ley de los Grandes N√∫meros: La probabilidad emp√≠rica converge a la te√≥rica a medida que N aumenta.")
    print("‚Ä¢ Teorema del L√≠mite Central: La distribuci√≥n de las sumas se aproxima a una distribuci√≥n normal.")
    print("‚Ä¢ An√°lisis de Convergencia: Visualizaci√≥n de c√≥mo el error disminuye con m√°s datos.")

# =============================================================================
# BLOQUE PRINCIPAL DE EJECUCI√ìN
# =============================================================================
if __name__ == "__main__":
    while True:
        ejecutar_simulacion()

        # Preguntar al usuario si desea realizar otra simulaci√≥n
        respuesta = input("\n¬øDesea realizar otra simulaci√≥n? (s/n): ").lower()
        if respuesta not in ['s', 'si', 'yes', 'y']:
            print("\n¬°Gracias por usar la simulaci√≥n del casino! ¬°Juegue con responsabilidad!")
            break

        print("\n" + "="*80)
        print("REINICIANDO SIMULACI√ìN...")
```
